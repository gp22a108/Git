```python
#課題1
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree

df = pd.read_csv('swfran.csv')

df_x = df[['length','left','right','bottom','top','diagonal']]
df_y = df['class']

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.3, random_state=1)

from sklearn import tree
model = tree.DecisionTreeRegressor(criterion='squared_error', max_depth=2)

model.fit(train_x, train_y)

pred = model.predict(test_x)
r2 = model.score(test_x, test_y)

print(pred)
print('\n')
print('R^2 =',r2)

from sklearn.tree import plot_tree
plt.figure(figsize=(12,10))
plot_tree(model, feature_names=train_x.columns, class_names=True, filled=True)
plt.show
```
