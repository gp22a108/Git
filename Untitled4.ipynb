```python
#課題1
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

# matplotlib の日本語表示
!pip install japanize-matplotlib
import japanize_matplotlib

# scikit-learn の主成分分析
from sklearn.decomposition import PCA

# データファイル読み込み
df = pd.read_csv('pilots.csv')
print(df)

# データの正規化
from scipy import stats
dfx = df.iloc[:, 1:].apply(stats.zscore, axis=0)
print('\n')
print(dfx)

# 主成分分析の実行
pca = PCA()
pca.fit(dfx)

# 主成分の分散
dfx_var = pd.DataFrame(pca.explained_variance_,
             columns = ["分散"],
             index = ["主成分{}".format(x+1) for x in range(len(dfx.columns))])
print("\n")
print(dfx_var)

# 主成分負荷量
dfx_pca_vector = pd.DataFrame(pca.components_,
                columns = [dfx.columns],
                index = ['主成分{}'.format(x+1) for x in range(len(dfx.columns))])
print('\n【主成分負荷量】')
print(dfx_pca_vector)

# 主成分得点
values = pca.transform(dfx)
dfx_pca_score = pd.DataFrame(data = values,
                columns = ['主成分{}'.format(x+1) for x in range(len(dfx.columns))])
print('\n【主成分得点】')
print(dfx_pca_score)

# 寄与率
dfx_ev_ratio = pd.DataFrame(data = pca.explained_variance_ratio_,
                columns = ['寄与率'],
                index = ['主成分{}'.format(x+1) for x in range(len(dfx.columns))])
print('\n 【寄与率】')
print(dfx_ev_ratio)
print('\n')

# 累積寄与率
cc_ev_ratio = np.cumsum(pca.explained_variance_ratio_)

# 0を連結
cc_ev_ratio = np.hstack([0, cc_ev_ratio])

# 寄与率をグラフ描画
plt.plot(cc_ev_ratio)
#plt.plot(cc_ev_ratio, "-o")
plt.xlabel('主成分')
plt.ylabel('累積寄与率')
plt.grid()
plt.show()
```

```python
#課題3(1)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# matplotlib の日本語表示
!pip install japanize-matplotlib
import japanize_matplotlib

from sklearn import model_selection, preprocessing, decomposition
from sklearn.decomposition import PCA

df = pd.read_csv('2021-23-food.csv')
print(df)

X = df.iloc[:, 1:12]
Y = df.iloc[:,0]
print(X)
print(Y)

# データの正規化
from scipy import stats
dfx = X.iloc[:, 0:].apply(stats.zscore, axis=0)
#print(dfx)

print('\n')
print('dfx shape:{}'.format(dfx.shape),'\n')

# 主成分分析の実行
pca = PCA(n_components=2)
pca.fit(dfx)

# データを主成分に変換
X_pca = pca.transform(dfx)

# 主成分得点
dfx_X_pca = pd.DataFrame(X_pca,
     columns =  ['主成分1', '主成分2'])
print(dfx_X_pca,'\n')

# 主成分負荷量
dfx_pca_components = pd.DataFrame(pca.components_,
    columns = [dfx.columns],
    index = ['主成分1', '主成分2'])
print(dfx_pca_components, '\n')

# 次元数
print('X_pca shape:{}'.format(X_pca.shape),'\n')

dfx_pca_explained_variance_ratio = pd.DataFrame(pca.explained_variance_ratio_,
    columns=['寄与率'],
    index = ['主成分1', '主成分2'])
print(dfx_pca_explained_variance_ratio, '\n')

# 主成分変量プロット
plt.figure(figsize=(6, 6))
for x, y, name in zip(pca.components_[0], pca.components_[1], dfx.columns[0:]):
    plt.text(x, y, name, ha='center', va='bottom')
plt.scatter(pca.components_[0], pca.components_[1])
plt.grid()
plt.xlabel("主成分1")
plt.ylabel("主成分2")
plt.show()

# 主成分得点プロット
plt.figure(figsize=(10,10))
plt.xlabel('主成分1')
plt.ylabel('主成分2')
plt.scatter(X_pca[:,0], X_pca[:,1], marker='.')
for i, label in enumerate(Y):
    plt.text(X_pca[i][0], X_pca[i][1], label, ha='center', va='bottom')
plt.show()
```
