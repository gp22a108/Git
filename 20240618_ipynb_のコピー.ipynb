```python
#課題1
import pandas as pd

#from sklearn.datasets import load_iris
df = pd.read_csv('customers_data.csv')
print(df)

df_x = df[[ 'Fresh', 'Milk', 'Grocery','Frozen','Detergents_Paper']]
df_y = df['Channel']

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x,df_y, test_size=0.2, random_state=1)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=15,max_features=5,random_state=1)
model.fit(train_x, train_y)

print(model.predict(test_x), '\n')
print(model.score(test_x,test_y))

import matplotlib.pyplot as plt

labels = train_x.columns
importances = model.feature_importances_

plt.figure(figsize = (4,2))
plt.barh(y = range(len(importances)), width = importances)
plt.yticks(ticks = range(len(labels)), labels = labels)
plt.show()
```

```python
#課題2
import pandas as pd

#from sklearn.datasets import fetch_california_housing
#housing = fetch_california_housing()
#housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
#housing_df['MedHouseVal'] = pd.DataFrame(housing.target)

df = pd.read_csv('diabetes.csv')

#print(housing_df)

df_x = df[['age', 'sex', 'bmi', 'bp','s2','s3']]
df_y = df['Disease']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=1)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=200, max_depth=10, max_features=8, random_state=1)
model.fit(x_train, y_train)

print(model.predict(x_test), '\n')
print(model.score(x_test, y_test))


from sklearn import tree
model_tree = tree.DecisionTreeRegressor(max_depth=10, random_state=1)
model_tree.fit(x_train, y_train)

print('決定木:',model_tree.score(x_test, y_test))
print('ランダムフォレスト:', model.score(x_test, y_test))
```

```python
#課題3
import pandas as pd

#from sklearn.datasets import fetch_california_housing
#housing = fetch_california_housing()
#housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
#housing_df['MedHouseVal'] = pd.DataFrame(housing.target)

df = pd.read_csv('auto-mpg.csv')

#print(housing_df)

df_x = df[['cylinders', 'horsepower', 'weight', 'acceleration','model-year','origin']]
df_y = df['mpg']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=1)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=20, max_depth=10, max_features=8, random_state=1)
model.fit(x_train, y_train)

print(model.predict(x_test), '\n')
print(model.score(x_test, y_test))


from sklearn import tree
model_tree = tree.DecisionTreeRegressor(max_depth=10, random_state=1)
model_tree.fit(x_train, y_train)

print('決定木:',model_tree.score(x_test, y_test))
print('ランダムフォレスト:', model.score(x_test, y_test))
```
